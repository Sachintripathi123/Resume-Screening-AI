```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
resume_df = pd.read_csv("/content/resumes_dataset.csv")
resume_df
```

```python
job_description_df = pd.read_csv("/content/job_descriptions_dataset.csv")
job_description_df
```

```python
job_description_df.info()
```

```python
def extract_skills_from_resume(resume_df):
  skills = []
  for skill in resume_df["skills"]:
     skills.append(skill)

  return skills
candidate_skills = extract_skills_from_resume(resume_df)
candidate_skills
```

```python
def extract_skills_from_job_description(job_description_df):
   prefered_job_id = int(input("Enter your Job Id which you want in requried skills : "))

   required_skills = job_description_df.loc[job_description_df["job_id"]==prefered_job_id,"required_skills"].tolist()
   return required_skills
required_skills = extract_skills_from_job_description(job_description_df)

required_skills
```

```python
def calculate_similarity(candidate_skills,required_skills):
  vector = TfidfVectorizer()
  corpus = candidate_skills+required_skills
  matrix = vector.fit_transform(corpus)
  simi = cosine_similarity(matrix)
  simi_df = pd.DataFrame(simi)
  score = (simi_df.iloc[-1,:-1])
  percentage =(score*100).tolist()
  percentage_df = pd.DataFrame({"Name":resume_df["name"],"percentage": percentage})
  return percentage_df
calculated_similarity = calculate_similarity(candidate_skills,required_skills)
calculated_similarity
```

```python
def rank_resume(percentage_df):
  ranked_df = percentage_df.sort_values(by="percentage",ascending = False)
  return ranked_df.reset_index(drop = True)
ranked_df = rank_resume(calculated_similarity)
ranked_df
```

```python
def display_result(ranked_df):
  print("="*30)
  print("RANKED RESUME")
  print("="*30)
  print(ranked_df.to_string(index=False))
display_result(ranked_df)
```

```python
def main():

    # Step 1
    candidate_skills = extract_skills_from_resume(resume_df)

    # Step 2
    required_skills = extract_skills_from_job_description(job_description_df)

    # Step 3
    calculated_similarity = calculate_similarity(
        candidate_skills,
        required_skills
    )

    # Step 4
    ranked_df = rank_resume(calculated_similarity)

    # Step 5
    display_result(ranked_df)
main()
```
