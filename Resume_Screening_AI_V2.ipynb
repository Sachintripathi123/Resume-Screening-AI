```python
import pandas as pd
import numpy as np
```

```python
resume_df = pd.read_csv('/content/resumes_dataset.csv')
resume_df
```

```python
job_df = pd.read_csv('/content/job_descriptions_dataset.csv')
job_df
```

```python
def extract_resume_skill_set(resume_df):
  resumes = resume_df["skills"]
  resume_skills_set = []

  for skill in resumes:
    skill = skill.split(',')
    cleaned_skill = []
    for s in skill:
      s = s.lower().strip()
      cleaned_skill.append(s)
    resume_set = set(cleaned_skill)
    resume_skills_set.append(resume_set)
  return resume_skills_set

resume_skills_set = extract_resume_skill_set(resume_df)
resume_skills_set
```

```python
def extract_job_skill_set(job_df):
  prefered_job_id = int(input("Enter your preferd Job ID : "))

  required_skills = job_df.loc[prefered_job_id==job_df["job_id"],"required_skills"].tolist()
  for skill in required_skills:
    skill = skill.split(",")
    cleaned_skill = []
    for s in skill:
      s = s.lower().strip()
      cleaned_skill.append(s)
    required_skills_set = set(cleaned_skill)
  return required_skills_set


required_skills_set = extract_job_skill_set(job_df)
required_skills_set
```

```python
def match_skills(resume_skills_set,required_skills_set):
  match_skills = []
  for matched_skill in resume_skills_set:
    matched_skill = matched_skill & required_skills_set
    match_skills.append(matched_skill)
  return match_skills
#resume_skills_set = extract_resume_skill_set(resume_df)

#required_skills_set = extract_job_skill_set(job_df)
match_skill = match_skills(resume_skills_set,required_skills_set)
match_skill
```

```python
def missing_skills(resume_skills_set,required_skills_set):
  missing_skills = []
  for resume_set in resume_skills_set:
    missed_skills  = required_skills_set - resume_set
    missing_skills.append(missed_skills)
  return missing_skills

missing_skill = missing_skills(resume_skills_set,required_skills_set)
missing_skill
```

```python
def skill_match_percentage(match_skills, required_skills_set):

    percentage = []

    for matched in match_skills:

        percent = (len(matched)/len(required_skills_set))*100

        percentage.append(percent)

    return percentage
resume_skills_set = extract_resume_skill_set(resume_df)

required_skills_set = extract_job_skill_set(job_df)


percentage = skill_match_percentage(match_skills(resume_skills_set, required_skills_set), required_skills_set)

print(percentage)
```

```python
def prepare_result(resume_df,
                   matched_skills,
                   missing_skills,
                   percentage):

    result_df = pd.DataFrame({
    "Name": resume_df["name"],
    "Matched Skills": matched_skills,
    "Missing Skills": missing_skills,
    "Match %": percentage
    })
    result_df = result_df.sort_values(by="Match %",ascending=False).reset_index(drop=True)

    return result_df
result_display = prepare_result(resume_df, match_skills(resume_skills_set, required_skills_set), missing_skills(resume_skills_set, required_skills_set), percentage)
result_display
```

```python
def main():

    # Step 1
    resume_skills_set = extract_resume_skill_set(resume_df)

    # Step 2
    required_skills_set = extract_job_skill_set(job_df)

    # Step 3
    match_skill = match_skills(
        resume_skills_set,
        required_skills_set
    )

    # Step 4
    missing_skill = missing_skills(
        resume_skills_set,
        required_skills_set
    )

    # Step 5
    percentage = skill_match_percentage(
        match_skill,
        required_skills_set
    )

    # Step 6
    result_display = prepare_result(
        resume_df,
        match_skill,
        missing_skill,
        percentage
    )

    return result_display
main()
```
